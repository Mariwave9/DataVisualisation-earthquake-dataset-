```python
#importing Pandas to load and read the csv file (dataset)
import pandas as pd
earthquake_dataset = pd.read_csv("all_month.csv")
earthquake_dataset.head()
```

```python
#analysing the structure of the dataset
earthquake_dataset.info()
```

```python
#generating descriptive statistics
earthquake_dataset.describe()
```

```python
#checking for missing values in the dataset, as some 'min' values in the descriptive statistics results were quite low
earthquake_dataset.isnull().sum()
```

```python
#importing matplotlib to generate the different visualisations to analise the dataset
import matplotlib.pyplot as plt
```

```python
#plotting an histogram to see how the data are distributed in the dataset
#histogram can be skewed -- positively skewed, goes towards the right positively skewed
earthquake_dataset.hist(bins= 10, figsize=(10,10))
plt.show()
```

```python
#replacing missing values with the median as the histogram above shows that the data are not normally distributed
#selected columns: mag, nst, gap, dmin, horizontalError, magError, magNst
earthquake_dataset['mag'] = earthquake_dataset['mag'].fillna(earthquake_dataset['mag'].median())
earthquake_dataset['nst'] = earthquake_dataset['nst'].fillna(earthquake_dataset['nst'].median())
earthquake_dataset['gap'] = earthquake_dataset['gap'].fillna(earthquake_dataset['gap'].median())
earthquake_dataset['dmin'] = earthquake_dataset['dmin'].fillna(earthquake_dataset['dmin'].median())
earthquake_dataset['horizontalError'] = earthquake_dataset['horizontalError'].fillna(earthquake_dataset['horizontalError'].median())
earthquake_dataset['magError'] = earthquake_dataset['magError'].fillna(earthquake_dataset['magError'].median())
earthquake_dataset['magNst'] = earthquake_dataset['magNst'].fillna(earthquake_dataset['magNst'].median())
```

```python
#checking the new dataset after replacing the missing values
earthquake_dataset.head()
```

```python
#plotting another histogram to see if the data appear differently with no missing values
earthquake_dataset.hist(bins= 10, figsize=(10,10))
plt.show()
```

```python
#Comparing both histograms we can see there is no significant difference between them when replacing the missing values.
#This small difference could be caused by the outliers present in the dataset, which can be checked creating a boxplot.
earthquake_dataset.boxplot(figsize=(20,20))
plt.show()
```

```python
#As can be seen in the boxplot above, there are a number of outliers, so removing or reducing their number could make a 
#difference in the further analysis of the data.

#removing outliers in latitude column
max_threshold, min_threshold = earthquake_dataset['latitude'].quantile([0.75,0.25])
earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.latitude>min_threshold) & (earthquake_dataset.latitude<max_threshold)]

#removing outliers in longitude column
max_threshold, min_threshold = earthquake_dataset['longitude'].quantile([0.75,0.25])
earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.longitude>min_threshold) & (earthquake_dataset.longitude<max_threshold)]

#removing outliers in depth column
max_threshold, min_threshold = earthquake_dataset['depth'].quantile([0.75,0.25])
earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.depth>min_threshold) & (earthquake_dataset.depth<max_threshold)]

#removing outliers in nst column
max_threshold, min_threshold = earthquake_dataset['nst'].quantile([0.75,0.25])
earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.nst>min_threshold) & (earthquake_dataset.nst<max_threshold)]

#removing outliers in gap column
max_threshold, min_threshold = earthquake_dataset['gap'].quantile([0.75,0.25])
earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.gap>min_threshold) & (earthquake_dataset.gap<max_threshold)]

#removing outliers in dmin column
max_threshold, min_threshold = earthquake_dataset['dmin'].quantile([0.75,0.25])
earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.dmin>min_threshold) & (earthquake_dataset.dmin<max_threshold)]

#removing outliers in horizontalError column
#max_threshold, min_threshold = earthquake_dataset['horizontalError'].quantile([0.75,0.25])
#earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.horizontalError>min_threshold) & (earthquake_dataset.horizontalError<max_threshold)]

#removing outliers in depthError column
max_threshold, min_threshold = earthquake_dataset['depthError'].quantile([0.75,0.25])
earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.depthError>min_threshold) & (earthquake_dataset.depthError<max_threshold)]

#removing outliers in magNst column
#max_threshold, min_threshold = earthquake_dataset['magNst'].quantile([0.75,0.25])
#earthquake_dataset_new = earthquake_dataset[(earthquake_dataset.magNst>min_threshold) & (earthquake_dataset.magNst<max_threshold)]

```

```python
earthquake_dataset_new.head()
```

```python
#checking out the outliers
earthquake_dataset_new.boxplot(figsize=(15,10))
plt.show()
```

```python
#importing seaborn to create a heatmap
import seaborn as sns

#Plotting a correlation matrix which displays the correlation coefficientes for the different variables will 
#depict the correlation between all the possible pairs of values in the table and help to identify and visualise 
#patterns in the given data.
corr_mat = earthquake_dataset_new.corr()
corr_mat
```

```python
#creating the heatmap
plt.figure(figsize=(10,10))
sns.heatmap(corr_mat, annot=True)
plt.title('Correlation Plot')
plt.show()
```

```python
#We can draw some conclusions from the heatmap above:
    ## nst (number of seismic stations) is quite related to the magnitude of the nst but inversely related to the gap
    ## longitude and latitude have negative and weak correlation
    ## magnitude is positively related to longitude but not latitude
    ## both latitude and longitude are inversely related to magError
```

```python
#counting the times each type of disaster occurred
earthquake_type = pd.DataFrame(earthquake_dataset['type'].value_counts())
earthquake_type.reset_index(inplace=True)
earthquake_type
```

```python
#Since the difference in the count is quite significant between types of disaster, I took into account the mean 
#of the magnitude of each type of disaster to analyse and compare the results
earthquaketype_mean = earthquake_dataset.groupby(['type']).mean('mag')['mag'].to_frame().reset_index()
earthquaketype_mean
```

```python
#creating a pie chart to visualise the mean of the magnitude of each type of disaster
fig, ax = plt.subplots()
earthquaketype_mean.plot(kind = 'pie', y = 'mag', labels = earthquaketype_mean['type'],
autopct= '%1.2f', ax=ax)
```

```python
#From this pie chart we can conclude that, although 'earthquake' is the most common type of disaster (with 9323 instances),
#'other events' have a higher average magnitude.
```

```python
#calculating the nst mean for each type of disaster
earthquaketype_mean_nst = earthquake_dataset.groupby(['type']).mean('nst')['nst'].to_frame().reset_index()
earthquaketype_mean_nst
```

```python
#creating a horizontal bar plot to see the difference in the nst mean between types of disaster
fig, ax = plt.subplots()
earthquaketype_mean_nst.plot(kind = 'barh', x = 'type', y = 'nst', color = 'red', ax=ax)
```

```python
#In the plot above, we can see that the number of seismic stations (nst) is higher in the type 'quarry blast'
```

```python
#extracting only the date from the 'time' variable in the dataset
earthquake_dataset['time'] = pd.to_datetime(earthquake_dataset['time']).dt.date
earthquake_dataset['time'].head()
```

```python
#calculating the nst mean for each day in the dataset
earthquake_time_nst = earthquake_dataset.groupby(['time']).mean('nst')['nst'].to_frame().reset_index()
earthquake_time_nst
```

```python
#creating a line chart to visualise the day(s) with the highest number of seismic stations (nst)
fig, ax = plt.subplots()
earthquake_time_nst.plot(figsize=(15,8),kind = 'line', x = 'time', y = 'nst', xlabel = 'date', color = 'purple', ax=ax)
```

```python
#calculating the magnitude mean for each day in the dataset
earthquake_time_mag = earthquake_dataset.groupby(['time']).mean('mag')['mag'].to_frame().reset_index()
earthquake_time_mag
```

```python
#creating a bar plot to visualise the day with the highest magnitude
fig, ax = plt.subplots()
earthquake_time_mag.plot(kind = 'bar', x = 'time', y = 'mag', color = 'orange', ax=ax)
```

```python
#From the above line chart and bar plot we can observe that the day with the highest number of seismic stations (nst) 
# and highest magnitude was the 21st of January. From a more scientific perspective it would be interesting to compare 
# this data to previous datasets to see if each month follows a similar pattern or whether it depends on different factors
# such as the season or place.
```
